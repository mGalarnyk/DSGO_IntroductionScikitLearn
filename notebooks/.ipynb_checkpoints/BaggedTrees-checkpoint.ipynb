<h2> What are bagged trees</h2> 

Each machine learning algorithm has strengths and weaknesses. A weakness of decision trees is that they are prone to overfitting on the training set. A way to mitigate this problem is to constrain how large a tree can grow. Bagged trees are an ensemble machine learning model. Ensemble models use multiple machine learning algorithms to hopefully obtain a better predictive performance than could be obtained from any individual model alone. Bagged trees try to mitigate the weaknesses of of decision trees by using bootstrapped data to grow multiple deep decision trees. The idea is that many trees protect each other from individual weaknesses.

## Tuning n_estimators (Number of Decision Trees)

A tuning parameter for bagged trees is <b>n_estimators</b>, which represents the number of trees that should be grown. 

```python
# List of values to try for n_estimators:
estimator_range = [1] + list(range(10, 150, 20))

scores = []

for estimator in estimator_range:
    reg = BaggingRegressor(n_estimators=estimator, random_state=0, oob_score = True)
    reg.fit(X_train, y_train)
    scores.append(reg.score(X_test, y_test))
```

```python
plt.figure(figsize = (10,7))
plt.plot(estimator_range, scores);

plt.xlabel('n_estimators', fontsize =20);
plt.ylabel('Score', fontsize = 20);
plt.tick_params(labelsize = 18)
plt.grid()
```

Notice that the score stops improving after a certain number of estimators (decision trees). One way to get a better score would be to include more features in the features matrix.

## Common questions

<h3>What is bootstrapped data</h3>
You can think of bootstrapped data as data generated from sample with replacement. Image taking a marble out of the bag and putting it back in. Shuffling the bag, and taking out another one. It is entirely possible you could take out the same marble that you have already taken out. 

```python
# In the code below it will generate a list of random numbers from 0 to 14
# it is WITH replacement. You will see duplicates.
# You will most likely not get the exact same numbers as me since it is random
np.random.choice(a=15, size=15, replace=True)
```

In the code below, note that you do not need to understand all of what the code does, only that a set in python can remove duplicates. If you are curious what they are, you can learn about them <a href="https://towardsdatascience.com/python-sets-and-set-theory-2ace093d1607">here</a>.

```python
# _ is a general purpose "throwaway" variable name to 
# indicate that the value produced by range is being deliberately ignored
# You will most likely not get the exact same numbers as me since it is random
for _ in range(1, 15):
    samples = np.random.choice(a=15, size=15, replace=True)
    print(samples, ',Num Samples: ' + str(len(samples)) + ',numberUnique: ' + str(len(set(samples))) )
```

```python
# You can also do this for a pandas dataframe
# Notice how you can have multiple identical rows
df.iloc[samples, :]
```

Note that about <a href="https://blogs.sas.com/content/iml/2017/06/28/average-bootstrap-sample-omits-data.html">2/3 of the data is unique.</a> For each tree in bagged tree or random forest models, the remaining observations that are not chosen from the sampling with replacement procedure are called "out-of-bag" observations.

The cool thing about this is that it allows you to estimate out of sample error without using model validation methods like train test split. 

How to calculate <b>"out-of-bag error":</b> 

1. For every observation in the training data, predict the target value using <b>only</b> the trees in which that observation was out-of-bag. Average predictions for regression or take a vote for classification.
2. Compare all predictions to the target values in order to compute the out-of-bag error.

```python
reg.oob_score_
```

<h3>What would happen if you didn't use bootstrapped data</h3>

If you didn't, you would have extremely similar decision trees defeating the purpose of averaging predictions.


"Decision trees are sensitive to the specific data on which they are trained. If the training data is changed (e.g. a tree is trained on a subset of the training data) the resulting decision tree can be quite different and in turn the predictions can be quite different."

<h3>What are some disadvantages of bagged trees</h3>

* They are less interpretable than decision trees. 
* Slower to train as you have n_estimators
* Bagging reduces the variance, but the trees can still be quite similar to each other. This is where random forests are useful.
